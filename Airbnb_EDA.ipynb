# Let's Read data from MongoDB

---


```python
from pymongo import MongoClient
import pandas as pd

mongodb = MongoClient("localhost",27017)
db = mongodb['Airbnb']
collection = db['airbnb_dataset']
v = collection.find({})
df = pd.DataFrame(v)
df
```

# **Let's Determine which columns are needed and drop the columns that are not needed**

```python
df.columns
```

```python
df.drop(['_id','listing_url', 'name', 'summary', 'space', 'description',
       'neighborhood_overview', 'notes', 'transit', 'access', 'interaction','minimum_nights','maximum_nights',
       'house_rules','bed_type', 'last_scraped', 'calendar_last_scraped', 'first_review', 'last_review',
        'bedrooms','bathrooms','amenities','security_deposit', 'cleaning_fee',
       'extra_people', 'guests_included', 'images', 'reviews', 'weekly_price',
       'monthly_price', 'reviews_per_month'],axis=1,inplace=True)
```

```python
df
```

# **Have to create few new columns to find insights**

```python
#extracting data from host column and creating new columns
df['Host_id'] = df['host'].apply(lambda x : x['host_id'])
df['Host_Name'] = df['host'].apply(lambda x : x['host_name'])

#extracting data from address column and creating new columns
df['Market'] = df['address'].apply(lambda x : x['market'])
df['Country'] = df['address'].apply(lambda x : x['country'])


#extracting data from availability column and creating new columns
df['Availability_365'] = df.availability.apply(lambda x : x['availability_365'])
```

```python
df
```

# **Looks like review scores has lots of '{}', lets write a function to extract the data we need and replace those '{}' with None**

```python
df['Review_score'] = df['review_scores'].apply(lambda x : x['review_scores_rating'] if x and ("review_scores_rating" in x) else None)
df['Country'] = df['Country'].str.replace('Hong Kong','China')
df['Country'] = df['Country'].str.replace('United States','USA')
```

```python
df
```

# **Lets again drop some unwanted columns**

```python
df.columns
```

```python
df.drop(['host','address','availability','review_scores'],axis=1,inplace=True)
```

# **Time to analyse the df.Review_score column and fill the null values(26.52% null values are there so lets fill it up)**

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.boxplot(df.Review_score)
```

**No need to eliminate the outliers as this is the review score**

**Lets fill the columns with median as this is a categorical and numerical value**

```python
df['Review_score'] = df['Review_score'].fillna(df.Review_score.median())
df['Review_score']
```

```python
sns.boxplot(df['Review_score'])
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```
